```python
#TODO: Aplicar gmm con hidden markov podria mejorar los datos artificiales
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from sklearn.mixture import GaussianMixture

class GMMTimeSeriesAugmentor:
    def __init__(self, num_components=5):
        self.num_components = num_components
        self.gmm = GaussianMixture(n_components=num_components)
    
    def fit(self, data):
        # Ajustar el GMM a los datos de entrada
        self.gmm.fit(data)
    
    def augment(self, num_samples):
        # Generar muestras adicionales usando el GMM
        augmented_samples = self.gmm.sample(num_samples)[0]
        return augmented_samples

# FunciÃ³n para visualizar las series de tiempo original y aumentada
def plot_original_and_augmented(original_series, augmented_series):
    plt.figure(figsize=(10, 5))
    plt.plot(original_series, label='Original', color='blue')
    plt.plot(augmented_series, label='Aumentada', color='red', linestyle='--')
    plt.xlabel('Tiempo')
    plt.ylabel('Valor')
    plt.title('Series de Tiempo Original y Aumentada')
    plt.legend()
    plt.show()

# Crear el DataFrame con tus datos de series de tiempo
# Supongamos que tu DataFrame se llama 'df' y contiene las series de tiempo en la columna 'cop_x'

# Ejemplo de uso
# Supongamos que df es tu DataFrame que contiene tus series de tiempo
# Extraemos las series de tiempo de la columna 'cop_x' y las convertimos en un array numpy
data = healthy_df = timeSeries[timeSeries['class'] == 'Neuropathic'].cop_x.values


# Convertir las listas de series de tiempo en arrays numpy
data = np.array([np.array(series) for series in data])

# Ajustar el augmentador a los datos
augmentor = GMMTimeSeriesAugmentor(num_components=1)
augmentor.fit(data)

# Generar una muestra adicional
num_samples = len(data)
augmented_data = augmentor.augment(num_samples)

# Visualizar la serie de tiempo original y la serie de tiempo aumentada
original_series = data[10]  # Tomamos la primera serie de tiempo del DataFrame
augmented_series = augmented_data[0]  # Tomamos la primera serie de tiempo aumentada
plot_original_and_augmented(original_series, augmented_series)
```

```python
import numpy as np

def pearson_correlation(series1, series2):
    # Calcula la covarianza entre las dos series
    covariance = np.cov(series1, series2)[0, 1]
    
    # Calcula las desviaciones estÃ¡ndar de cada serie
    std_dev1 = np.std(series1)
    std_dev2 = np.std(series2)
    
    # Calcula el coeficiente de correlaciÃ³n de Pearson
    correlation = covariance / (std_dev1 * std_dev2)
    
    return correlation

def series_membership(series, set_of_series, threshold):
    # Inicializa una lista para almacenar las correlaciones
    correlations = []
    
    # Calcula la correlaciÃ³n de Pearson entre la serie dada y cada serie en el conjunto
    for series_in_set in set_of_series:
        correlation = pearson_correlation(series, series_in_set)
        correlations.append(correlation)
    
    # Verifica si la serie dada tiene una correlaciÃ³n suficientemente alta con al menos una serie en el conjunto
    is_member = any(correlation > threshold for correlation in correlations)
    
    return is_member

# Ejemplo de uso
# Supongamos que 'series' es la serie de tiempo dada y 'set_of_series' es el conjunto de series de tiempo
# y 'threshold' es el umbral de correlaciÃ³n que se utilizarÃ¡ para determinar la pertenencia

# Definimos el umbral de correlaciÃ³n
threshold = 0.6
data1 = healthy_df = timeSeries[timeSeries['class'] == 'Healthy'].cop_x.values
data2 = healthy_df = timeSeries[timeSeries['class'] == 'Diabetic'].cop_x.values
data3 = healthy_df = timeSeries[timeSeries['class'] == 'Neuropathic'].cop_x.values


print("Â¿La serie dada pertenece al conjunto1?:", series_membership(augmented_data[29], data1, threshold))
print("Â¿La serie dada pertenece al conjunto?2:", series_membership(augmented_data[29], data2, threshold))
print("Â¿La serie dada pertenece al conjunto?:3", series_membership(augmented_data[29], data3, threshold))
#TODO: Hacer un filtro para eliminar las series que no sean del mismo tipo de clase o que pertenecen a varias clases.

```
