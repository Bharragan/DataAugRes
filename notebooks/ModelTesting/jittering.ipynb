```python
import numpy as np, sys,os
import pandas as pd
import matplotlib.pyplot as plt

import sys
sys.path.insert(0, "..")
from src.visualization.visualize import plot_original_and_augmented

from src.visualization.visualize import plot_series

timeSeries = pd.read_json('../data/processed/realData.json', orient='records', lines=True)
```

```python
import numpy as np

def add_gaussian_noise(time_series, sigma=0.03):
    """
    Aade ruido gaussiano a una serie de tiempo.

    Parameters:
    time_series (list or np.ndarray): La serie de tiempo original.
    sigma (float): La desviacin estndar del ruido gaussiano. Por defecto es 0.03.

    Returns:
    np.ndarray: La serie de tiempo con ruido gaussiano aadido.
    """
    # Asegurarse de que la serie de tiempo sea un array numpy
    time_series = np.array(time_series)
    
    return time_series + np.random.normal(loc=0., scale=sigma, size=time_series.shape)



def augment_time_series_with_noise(X):
    """
    Augmenta un conjunto de series de tiempo aadiendo ruido gaussiano.

    Parameters:
    X (list of lists or np.ndarray): Conjunto de series de tiempo originales.

    Returns:
    augmented_data (list): Conjunto de series de tiempo con ruido gaussiano aadido.
    """
    # Inicializar lista para almacenar las series de tiempo aumentadas
    augmented_data = []
    
    # Iterar sobre cada serie de tiempo en el conjunto original
    for series in X:
        
        # Aadir ruido gaussiano a la serie de tiempo
        noisy_series = add_gaussian_noise(series)
        
        # Agregar la serie de tiempo con ruido a la lista
        augmented_data.append(noisy_series)
    
    return augmented_data

# Ejemplo de uso con un conjunto de series de tiempo
data = timeSeries.cop_x

noisy_time_series_set = augment_time_series_with_noise(data)
plot_series(data, noisy_time_series_set, plot_original_first=True)
```
